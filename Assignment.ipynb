# Task 5

CEMA works by observing the physical state (power) of the chip and analyzing it statistically to pick the most suited key. While brute forcing follows a "all or nothing" sort of binary approach where you try a key and if it doesn't work, it's discarded till you find the correct one, CEMA's accuracy increases with the number of tests we run on the device, where at some point you will find the key. 

In this demo we attempt to find a key with 16 bytes. One byte is 8 bits. Which means there are 16 x 8 = 128 bits in the key. Meaning, there are 2^128 possibilities to check. It is said that cracking a password with that many keys (by brute force) would take 1 billion billion years.

When brute forcing, assuming we know the key length and the crypto implementation, the only inputs we have are the original text and the text after an encryption. We can then try all possible combinations for the key with the original text to see what matches the encrypted text.

But in CEMA, we have more insights into what is going on, we have power variations across the chip as a source of data too. Because we can sort of observe a correlation between the input (plain text) and the encryption. We can try many inputs and observe the patterns of power in the encrypted texts. Then, statistically we can narrow down on the key.
Of course, all of this is under the assumption that we can put the embedded device in a mode where it will perform encryptions to any amount of plain texts provided as input.
